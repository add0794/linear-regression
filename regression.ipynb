```python
import os
import pandas as pd
import IPython
```

```python
# Creating a Pandas dataframe of the data

cwd = os.getcwd()

df = pd.read_csv("cost_revenue_dirty.csv")
cols = list(df.columns)
print(df.shape)
df.head()
```

```python
# Preparing/Cleaning Data


# 1) Renaming 

df = df.rename(columns={"Production Budget ($)": "production_budget", "Worldwide Gross ($)": "worldwide_gross_usd", "Domestic Gross ($)": "domestic_gross_usd"})


# 2) Removing rows with $0

columns_to_strip = ["production_budget", "worldwide_gross_usd", "domestic_gross_usd"]

for col in columns_to_strip:
    df[col] = df[col].str.strip("$")


# #3) Removing null/missing data

df.isnull().sum()
df.dropna(inplace=True)


# 4) Removing duplicates

df.drop_duplicates(inplace=True)


df.head()
```
